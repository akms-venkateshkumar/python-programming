# L2 — Jupyter Notebook Example

This notebook demonstrates how to structure a small exploratory analysis in notebook cells. It uses seaborn's `tips` dataset so the notebook runs without external data.

## Imports and dataset load

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(style="whitegrid")
df = sns.load_dataset("tips")
print("Dataset shape:", df.shape)
```

## Quick peek

```python
df.head()
```

## Inline plot (visualization)

```python
plt.figure(figsize=(8,5))
sns.histplot(df['total_bill'], bins=20, kde=True)
plt.title("Total bill distribution")
plt.xlabel("Total bill ($)")
plt.show()
```

## Derived metric and summary table

```python
df['tip_pct'] = (df['tip']/df['total_bill']) * 100
df.groupby('day')['tip_pct'].agg(['mean','median','count']).round(2)
```

## Save results

```python
df.to_csv("tips_with_tip_pct.csv", index=False)
print("Saved tips_with_tip_pct.csv")
```
