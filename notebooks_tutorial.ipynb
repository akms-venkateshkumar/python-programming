## 1) Setup: imports and load data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(style="whitegrid")
df = sns.load_dataset("tips")
df.shape
```

### What to notice:
- The dataset includes columns such as `total_bill`, `tip`, `sex`, `smoker`, `day`, `time`, and `size`.
- Small datasets like this are ideal for learning — in real projects you'll load CSVs or databases.


---

## 2) Inspecting the data

```python
# Quick overview
df.head()
```

```python
# Dtypes and missing values
df.info()
print("")
print("Missing per column:")
print(df.isna().sum())
```

---

## 3) Cleaning & transforming
Show simple examples: rename, handle missing values, enforce categories.


```python
# Standardize column names and fill missing tip values (if any)
df = df.rename(columns=str.strip).copy()
df['tip'] = df['tip'].fillna(df['tip'].median())
for col in ['sex','smoker','day','time']:
    df[col] = df[col].astype('category')
df.dtypes
```

---

## 4) Analysis: derived metrics and grouping


```python
# Create a derived metric: tip percentage
df['tip_pct'] = (df['tip'] / df['total_bill']) * 100
# Which day/time has the highest average tip percent?
df.groupby(['day','time'])['tip_pct'].mean().round(2).unstack()
```

```python
# A couple of quick plots to visualize results
plt.figure(figsize=(8,5))
sns.boxplot(x='day', y='tip_pct', data=df)
plt.title('Tip percentage by day')
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(x='total_bill', y='tip_pct', hue='time', data=df)
plt.title('Tip % vs Total bill')
plt.show()
```

---

## 5) Communicating results
Save a small CSV and an image to share with stakeholders. Include a short textual summary.


```python
# Save outputs
df.to_csv('tutorial_tips_with_tip_pct.csv', index=False)
plt.figure(figsize=(8,5))
sns.boxplot(x='day', y='tip_pct', data=df)
plt.title('Tip percentage by day')
plt.savefig('tutorial_tip_pct_by_day.png')
plt.close()
print('Wrote CSV and PNG')
```
